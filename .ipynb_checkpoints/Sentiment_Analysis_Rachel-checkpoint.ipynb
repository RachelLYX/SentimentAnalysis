```python
import pandas as pd
import re
```

```python
df = pd.read_csv("final_dataset _HP_manas.csv")
df
```

```python
df.isnull().sum()
```

```python
country_map = {
    "united states": "US",
    "italy": "IT",
    "spain": "ES",
    "germany": "DE",
    "france": "FR",
    "united kingdom": "GB",
    "canada": "CA",
    "mexico": "MX"
}

df.loc[:, "countryCode"] = df["country"].str.strip().str.lower().map(country_map).fillna(df["countryCode"])
```

```python
df.head(62)
```

```python
text_cols = [
    "reviewDescription", "reviewTitle", "reviewUrl", "variant", "productAsin"
] + [col for col in df.columns if "reviewImages" in col]

df[text_cols] = df[text_cols].fillna("NA")

num_cols_zero = ["reviewReaction", "position"]
df[num_cols_zero] = df[num_cols_zero].fillna(0)

df["ratingScore"] = df["ratingScore"].fillna(df["ratingScore"].median())

df["isVerified"] = df["isVerified"].fillna(False)

if df["date"].isna().any():
    df["date"] = df["date"].fillna(pd.NaT)
```

```python
def extract_helpful_count(text):
    if pd.isna(text):
        return 0
    text = str(text).lower().strip()
    if text == "0":
        return 0
    if "one person" in text:
        return 1
    match = re.search(r"(\d+)\s+people", text)
    if match:
        return int(match.group(1))
    return 0

df["reviewReaction"] = df["reviewReaction"].apply(extract_helpful_count)
```

```python
df.to_csv("cleaned_dataset.csv", index=False)
```
